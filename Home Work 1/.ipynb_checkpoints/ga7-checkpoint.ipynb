```python
fig, ax= plt.subplots(1,2)
df[df['Sex'] == 'female'].plot('Age','Fare', kind = 'Scatter', ax =[0], color = 'red', label = 'Females')
df[df['Sex'] == 'male'].plot('Age','Fare', kind = 'Scatter', ax =[1], color = 'red', label = 'Males')
```
